4. Create a line plot that shows the number of talks released every year between 1972 and 2017.

```python
data_copy = data.copy(deep = True)
data_copy['published_year'] = pd.to_datetime(data_copy['published_date'], unit='s').apply(lambda x : x.year)
```

```python
data_copy['published_year'].unique()
```

```python
n_talks = data_copy.groupby('published_year').size().reset_index(name = 'count')
```

```python
n_talks
```

```python
plt.figure(figsize = (8, 6))
plt.plot(n_talks['published_year'], n_talks['count'], marker='o', linestyle='-', color='b')
plt.title('Yearwise count')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(ticks = n_talks['published_year'])
plt.show()
```

5. Create a heatmap that shows the number of talks that have taken place each month since 1972.

```python
data_copy2 = data.copy(deep = True)
data_copy2['film_date'] = pd.to_datetime(data_copy['film_date'], unit='s')
data_copy2['film_year'] = data_copy2['film_date'].apply(lambda x : x.year)
data_copy2['film_month'] = data_copy2['film_date'].apply(lambda x : x.month)
```

```python
monthly_talks = data_copy2.groupby(['film_year', 'film_month']).size().reset_index(name = 'count')
```

```python
heatmap_data = monthly_talks.pivot(index='film_year', columns='film_month', values='count').fillna(0)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', linewidths=0.5, linecolor='gray')
plt.title('Number of TED Talks per Month (1972 onwards)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.xticks(ticks=[i-1 for i in range(1, 13)], labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

6. Are speakers from certain professions more likely to get more views? Demonstrate using a graph.

```python
data_occupation = data.groupby('speaker_occupation')['views'].mean().reset_index(name = 'avg_views')
data_occupation.sort_values(by = 'avg_views', inplace = True)
```

```python
top_professions = data_occupation.tail(10)
```

```python
top_professions
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='avg_views', y='speaker_occupation', hue='speaker_occupation', data=top_professions, palette='viridis', legend = False)
plt.title('Top 10 Professions by Average TED Talk Views', fontsize=16)
plt.xlabel('Average Views', fontsize=12)
plt.ylabel('Profession', fontsize=12)
plt.show()
```

7. Identify the top themes of TED Talks. Next, show the trends in the number of talks across themes over the years.

```python
data_copy3 = data.copy(deep = True)
data_copy3['tags'] = data_copy3['tags'].str.split(',')
data_exploded = data_copy3.explode('tags')
data_exploded['tags'] = data_exploded['tags'].str.replace(r'[\[\]]', '', regex=True).str.strip()
```

```python
data_exploded
```

```python
data_themes = data_exploded.groupby('tags').size().reset_index(name = 'count')
data_themes.sort_values(by = 'count', ascending = False, inplace = True)
data_themes = data_themes[~data_themes['tags'].str.contains('TED', case=False, na=False)]
```

```python
print('The top 10 themes are: ')
print()
for theme in data_themes['tags'].iloc[:10]:
    print(theme)
```

```python
data_exploded['film_year'] = pd.to_datetime(data_exploded['film_date'], unit='s').dt.year
theme_trends = data_exploded.groupby(['tags', 'film_year']).size().reset_index(name='count')
theme_trends = theme_trends[theme_trends['tags'].isin(data_themes['tags'].iloc[:5])]
```

```python
theme_trends
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=theme_trends.sort_values(by = ['tags', 'film_year']), x='film_year', y='count', hue='tags')

plt.title('Trends in the Number of TED Talks Across top 5 Themes Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Talks')
plt.legend(title='Theme', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
fig = px.line(theme_trends, x='film_year', y='count', color='tags',
              title='Trends in the Number of TED Talks Across top 5 Themes Over the Years',
              labels={'film_year': 'Year', 'count': 'Number of Talks', 'tags': 'Theme'})

# Show the plot
fig.show()
```

8. What words are most frequently used in TED Talks? (discounting common stopwords such as the, and, a, etc.). Create a wordcloud demonstrating word usage in TED Talks (you may need to use a library outside of matplotlib and seaborn for this).

```python
transcript_data = pd.read_csv(r"C:\Personal\The Scrappy Project\module_5\transcripts.csv")
```

```python
all_text = ' '.join(transcript_data['transcript'].dropna())
```

```python
stopwords = set([
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself', 'yourselves',
    'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their',
    'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are', 'was',
    'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and',
    'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between',
    'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off',
    'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any',
    'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so',
    'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'should', 'now'
])
```

```python
wordcloud = WordCloud(stopwords=stopwords, width=800, height=400, background_color='white').generate(all_text)

# Plot the word cloud
plt.figure(figsize=(16, 16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of TED Talks')
plt.show()
```
